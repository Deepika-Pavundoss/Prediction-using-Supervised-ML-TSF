Linear Regression is a supervised machine learning algorithm where the predicted output is continuous straight line having constant slope. 

The predicted result has equation of the form:  y = mx + c

# Importing libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Importing the dataset


```python
dataset = pd.read_csv("http://bit.ly/w-data")
dataset
```

```python
x=dataset.iloc[:,:-1].values
```

```python
x
```

```python
y = dataset.iloc[:,-1].values
y
```

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0);
```

# Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

#Predicting the Test set results


```python
y_pred = regressor.predict(x_test)
y_pred
```

#Visualizing the Predicted Results


```python
m = regressor.coef_
c = regressor.intercept_
y_line = m*x_train+c
```

```python
plt.plot(x_train,y_line,color = 'blue')
plt.scatter(x_train,y_train,color='red') 
plt.scatter(x_test,y_pred,color='green') #predicted
plt.title('Linear Regression')
plt.xlabel('No of study hours')
plt.ylabel('Percentage of Student')
plt.show()
```

# Predicting a Single Result

```python
print(regressor.predict([[9.25]]))
```

# Evaluating

```python
results = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
results
```

```python
from sklearn import metrics  
metrics.mean_absolute_error(y_test, y_pred)
```
